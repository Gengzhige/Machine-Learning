## 绘制plot曲线

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
x
```

```python
y = x * x
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-10, 10, 5)
y = x * x
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0,10,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y1, color='green')
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
plt.xlim(-5, 15)
plt.ylim(-2, 2)
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
plt.axis([-5,15,-2,2])
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':',label='sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':',label='sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title('This is my first plot')
plt.show()
```
